```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

dataset = [['Bread', 'Jelly', 'PeanutButter'],
           ['Bread', 'PeanutButter'],
           ['Bread', 'Milk', 'PeanutButter'],
           ['Beer', 'Bread'],
           ['Beer', 'Milk']]

te = TransactionEncoder()
te_setResult = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_setResult, columns=te.columns_)

result = apriori(df, min_support = 0.5, use_colnames = True)
print(result)
print('\n')
association_rules(result, metric="confidence", min_threshold=0.1) 
```
